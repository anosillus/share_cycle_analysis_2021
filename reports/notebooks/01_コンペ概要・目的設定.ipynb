#### 所見2(データを見た後)

- 課題設定を間違えているクソコンペ

- スタッフの気紛れに依存せすに、確実に予測できるのは、「ステーションの配備可能台数よりもバイクの需要が多いから、バイクの残り台数は0になる」のときだけ。
